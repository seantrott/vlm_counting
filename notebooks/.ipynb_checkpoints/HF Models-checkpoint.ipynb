## Using Open Models



```python
import json
import base64
import pandas as pd
import seaborn as sns

from PIL import Image
from tqdm import tqdm


### Transformers
from transformers import ViltProcessor, ViltModel, ViltForQuestionAnswering
import torch
import torch.nn.functional as F
```

## Loading stimuli

```python
with open("paths.json", "r") as f:
    paths = json.load(f)
```

```python
# Function to encode the image
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
responses = []

for num_dots, img_name in tqdm(paths.items()):

    ### encode image
    img = encode_image(img_name)
```

## Loading models

```python
MODELS = ['dandelin/vilt-b32-finetuned-vqa',
          'dandelin/vilt-b32-mlm']
m = MODELS[0]
```

```python
# Load the processor and model
processor = ViltProcessor.from_pretrained(m)
model = ViltForQuestionAnswering.from_pretrained(m)
```

```python
# Define the question
question = "How many dots are in the image? Please answer with a single integer number."
```

```python
# Oepn image
print(img_name)

image = Image.open(img_name).convert("RGB")
image
```

```python
# Process the inputs
inputs = processor(images=image, text=question, return_tensors="pt")
```

```python
# Perform a forward pass
outputs = model(**inputs)

# Extract the logits
logits = outputs.logits

# Get the predicted answer ID
answer_id = logits.argmax(-1).item()

# Load the VQA answer list (assuming the model is trained on the VQA dataset)
vqa_answers = processor.tokenizer.convert_ids_to_tokens(range(processor.tokenizer.vocab_size))

# Get the predicted answer
predicted_answer = vqa_answers[answer_id]

print(f"Predicted answer: {predicted_answer}")
```

```python
processor.tokenizer.convert_ids_to_tokens(answer_id)
```

```python
logits.shape
```

```python
# Apply softmax to get probabilities
probabilities = F.softmax(logits, dim=-1)
```

```python
# Find the token ID for the specific token "2"
token_id = processor.tokenizer.convert_tokens_to_ids("9")

# Get the probability for the token "2"
probability_of_2 = probabilities[0, token_id].item()

print(f"Probability of token '2': {probability_of_2}")
```
